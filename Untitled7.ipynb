```python
import pandas as pd
from sklearn.model_selection import train_test_split

# 데이터 불러오기
data = pd.read_csv('/content/DATA.txt', sep = "\t")


# 독립 변수 (Year)와 종속 변수 (Price1, Price2, Price3) 설정
X = data['Year']  # 공통 독립 변수는 Year입니다.
y1 = data['Price1']  # Price1 예측용 종속 변수
y2 = data['Price2']  # Price2 예측용 종속 변수
y3 = data['Price3']  # Price3 예측용 종속 변수

# 80% 훈련 데이터, 20% 테스트 데이터로 분리
X_train1, X_test1, y_train1, y_test1 = train_test_split(X, y1, test_size=0.2, random_state=42)
X_train2, X_test2, y_train2, y_test2 = train_test_split(X, y2, test_size=0.2, random_state=42)
X_train3, X_test3, y_train3, y_test3 = train_test_split(X, y3, test_size=0.2, random_state=42)

# 데이터 크기 확인
print(f"훈련 데이터 크기 (Price1): {X_train1.shape}, 테스트 데이터 크기: {X_test1.shape}")
print(f"훈련 데이터 크기 (Price2): {X_train2.shape}, 테스트 데이터 크기: {X_test2.shape}")
print(f"훈련 데이터 크기 (Price3): {X_train3.shape}, 테스트 데이터 크기: {X_test3.shape}")

```

```python
print(type(X_test3))  # X_train1의 데이터 타입 확인
print(type(y_test3))  # y_train1의 데이터 타입 확인

import numpy as np

# X_train1이 DataFrame 또는 numpy 배열이 아닌 경우 변환
if not isinstance(X_test3, (pd.DataFrame, np.ndarray)):
    X_test3 = pd.DataFrame(X_test3)

# y_train1도 마찬가지로 변환
if not isinstance(y_test3, (pd.Series, np.ndarray)):
    y_test3 = pd.Series(y_test3)
X_test3 = np.array(X_test3)
y_test3 = np.array(y_test3)

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train3 = scaler.fit_transform(X_train3)
X_test3= scaler.transform(X_test3)
```

```python
print(type(X_test1))  # X_train1의 데이터 타입 확인
print(type(y_test1))  # y_train1의 데이터 타입 확인

import numpy as np

# X_train1이 DataFrame 또는 numpy 배열이 아닌 경우 변환
if not isinstance(X_test1, (pd.DataFrame, np.ndarray)):
    X_test1 = pd.DataFrame(X_test1)

# y_train1도 마찬가지로 변환
if not isinstance(y_test1, (pd.Series, np.ndarray)):
    y_test1 = pd.Series(y_test1)
X_test1 = np.array(X_test1)
y_test1 = np.array(y_test1)

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train1 = scaler.fit_transform(X_train1)
X_test1 = scaler.transform(X_test1)
```

```python
from flaml import AutoML

# Price1 예측용 모델 훈련
automl1 = AutoML()
automl1.fit(X_train1, y_train1, task="regression", time_budget=300, metric='r2', log_file_name="automl1.log", n_jobs=-1)

# Price2 예측용 모델 훈련
automl2 = AutoML()
automl2.fit(X_train2, y_train2, task="regression", time_budget=300, metric='r2', log_file_name="automl2.log", n_jobs=-1)

# Price3 예측용 모델 훈련
automl3 = AutoML()
automl3.fit(X_train3, y_train3, task="regression", time_budget=300, metric='r2', log_file_name="automl3.log", n_jobs=-1)

# 최적의 모델 출력
print(f"Best model for Price1: {automl1.model.estimator}")
print(f"Best model for Price2: {automl2.model.estimator}")
print(f"Best model for Price3: {automl3.model.estimator}")
```

```python
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

# Price1 예측 및 성능 평가
y_pred1 = automl1.predict(X_test1)
r2_1 = r2_score(y_test1, y_pred1)
rmse_1 = np.sqrt(mean_squared_error(y_test1, y_pred1))
print(f"Price1 - R2 Score: {r2_1:.4f}, RMSE: {rmse_1:.4f}")

# Price2 예측 및 성능 평가
y_pred2 = automl2.predict(X_test2)
r2_2 = r2_score(y_test2, y_pred2)
rmse_2 = np.sqrt(mean_squared_error(y_test2, y_pred2))
print(f"Price2 - R2 Score: {r2_2:.4f}, RMSE: {rmse_2:.4f}")

# Price3 예측 및 성능 평가
y_pred3 = automl3.predict(X_test3)
r2_3 = r2_score(y_test3, y_pred3)
rmse_3 = np.sqrt(mean_squared_error(y_test3, y_pred3))
print(f"Price3 - R2 Score: {r2_3:.4f}, RMSE: {rmse_3:.4f}")

```

```python
# 2023년도 가격 예측
year_2023 = np.array([[2023]])

# Price1 예측
predicted_price1_2023 = automl1.predict(year_2023)
# Price2 예측
predicted_price2_2023 = automl2.predict(year_2023)
# Price3 예측
predicted_price3_2023 = automl3.predict(year_2023)

# 결과 출력
print(f"Predicted prices for 2023:")
print(f"Price1: {predicted_price1_2023[0]:.2f}")
print(f"Price2: {predicted_price2_2023[0]:.2f}")
print(f"Price3: {predicted_price3_2023[0]:.2f}")
```
